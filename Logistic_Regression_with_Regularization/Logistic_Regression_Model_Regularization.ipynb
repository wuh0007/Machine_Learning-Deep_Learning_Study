```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import numpy as np # linear algebra
import math

# Input data files are available in the same directory as the .py file
# Any results you write to the current directory are saved as output
import os
tmp = os.listdir(r"C:\Users\wuh00\Machine_Learning_Study\Logistic_Regression_with_Regularization")
[x.encode('utf-8') for x in tmp]

# Load the libraries and data...
import matplotlib.pyplot as plt
import matplotlib.animation as animation

data = pd.read_csv('Sample Data.txt',header=None, index_col=False)
# print(data)


# define functions
def normalize_features(x):
    m = len(x) #  find size of the sample = m
    
    #  formula implementation x_norm = m x n, x = m x n, mean(x) and std(x), 
    #  which are 1 x n, need to be m x n, use np.tile
    x_norm = (x - np.tile(x.mean(),[m,1])) / np.tile(x.std(),[m,1])     
    return x_norm

def compute_sigmoid(z):
    # Compute sigmod function
    # Initialize the g(z)
    g = np.zeros(z.shape)
    # sigmoid formula 
    g = 1 / ( 1 + np.exp(-z))   
    return g

def compute_cost_logistic_regression_regularized(theta,x_norm,y,lamta):
    # Compute cost for logistic regression 
    [m,n] = x_norm.shape  # number of training examples
    
    # Compute cost
    h = compute_sigmoid(np.dot(x_norm,theta)) 
    J = -(1/m)*np.sum(y * np.log(h) + (1 - y) * np.log(1 - h)) + (lamta/(2*m)) * np.sum(np.square(theta[1:]))    
    return J

def gradient_descent_logistic_regression_regularized(theta,x_norm,y,alpha,num_iters,lamta):
    # gradient descent for logistic regression 
    [m,n] = x_norm.shape    
    past_costs = []
    for idx in range(num_iters):  
        # final theta vector 
        # theta(0)
        theta[0] = theta[0] - (alpha/m) * np.dot((compute_sigmoid(np.dot(x_norm,theta))-y).transpose(),x_norm[:,0]).transpose()
        # theta(1-j)                             
        theta[1:] = np.dot(theta[1:],(1-((alpha*lamta)/m))) - ((alpha/m) * np.dot((compute_sigmoid(np.dot(x_norm,theta))-y).transpose(),x_norm[:,1:]).transpose())
        J = compute_cost_logistic_regression_regularized(theta,x_norm,y,lamta)   
        past_costs.append(J)
    return theta,past_costs

# def remove_values_from_list(the_list, val):
#     return [value for value in the_list if value != val]

def performance_measure(predicted_y, actual_y):
    # performance measurement for logistic regression 
    # total number of samples
    m = actual_y.shape[0]    
    # number of samples(h=y)
    
    # number of samples(h=y)/total number of samples
    overall_accuracy =  np.mean(predicted_y == actual_y) * 100
    # number of samples(h=y=1)
    tp_count = len(np.nonzero((actual_y==1) & (predicted_y==1))[0])
    # number of samples(h=1,y=0)
    fp_count = len(np.nonzero((actual_y==0) & (predicted_y==1))[0])
    
    return overall_accuracy,tp_count,fp_count

# Target values, last column of the data frame
y = data.iloc[:,-1]

# feature values, all the columns except the last column
x = data.iloc[:,0:-1]

x1 = data.loc[y == 1]
x2 = data.loc[y == 0]
# print(x1)
# print(x2)
plt.xlabel('Feature x1')
plt.ylabel('Feature x2')
plt.plot(x1.iloc[:,0],x1.iloc[:,1],'bo',label='Negative Class')
plt.plot(x2.iloc[:,0],x2.iloc[:,1],'ro',label='Positive Class')
#     plt.scatter(admitted.iloc[:, 0], admitted.iloc[:, 1], s=10, label='Admitted')
#     plt.scatter(not_admitted.iloc[:, 0], not_admitted.iloc[:, 1], s=10, label='Not Admitted')
plt.legend(loc='upper right')
plt.show()

x_norm = normalize_features(x)
m = x_norm.shape[0] 
x_ones = np.ones((m,1))
x_norm = np.concatenate((x_ones,x_norm),axis=1)

y = y[:, np.newaxis]

# other parameters
n = x_norm.shape[1]
theta = np.zeros((n,1))
alpha = 0.1   #learning rate
iterations = 3000   #No. of iterations
lamta = 10

#Pass the relevant variables to the function and get the new values back...
theta,past_costs = gradient_descent_logistic_regression_regularized(theta,x_norm,y,alpha,iterations,lamta)

#Print the results...
print("Gradient Descent: ",theta[0], theta[1],theta[2])

#Plot the cost function...
plt.title('Cost Function J')
plt.xlabel('No. of iterations')
plt.ylabel('Cost')
plt.plot(past_costs)
plt.show()

h = compute_sigmoid(np.dot(x_norm,theta))
[overall_accuracy,tp_count,fp_count] = performance_measure((h>0.5).astype(int), y.astype(int))
print("overall accuracy: ",overall_accuracy)
print("True points count: ",tp_count)
print("false points count: ",fp_count)

x3 = data.loc[np.nonzero((h>0.5).astype(int) == 0)[0]]
x4 = data.loc[np.nonzero((h>0.5).astype(int) == 1)[0]]
plt.xlabel('Feature x1')
plt.ylabel('Feature x2')
plt.plot(x1.iloc[:,0],x1.iloc[:,1],'bo',label='Negative Class - True Label')
plt.plot(x2.iloc[:,0],x2.iloc[:,1],'ro',label='Positive Class - True Label')
plt.plot(x3.iloc[:,0],x3.iloc[:,1],'b*',label='Negative Class - Predicted Label')
plt.plot(x4.iloc[:,0],x4.iloc[:,1],'r*',label='Positive Class - Predicted Label')
#     plt.scatter(admitted.iloc[:, 0], admitted.iloc[:, 1], s=10, label='Admitted')
#     plt.scatter(not_admitted.iloc[:, 0], not_admitted.iloc[:, 1], s=10, label='Not Admitted')
plt.legend(loc='upper right')
plt.show()
```
